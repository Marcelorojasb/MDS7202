````python
# Se realiza un melt sobre los datos de las columnas de score y asi poder definir a las asignaturas como categorias
df_melt=df_grades[["math score","reading score","writing score"]].melt(var_name="subject")

# Se genera el box plot
fig = px.box(df_melt, y="value", x="subject", color="subject",labels={"value":"score"})
fig.update_xaxes(showticklabels=False, title=None) # Elimina las etiquetas y títulos en el eje x

fig.show()
````

**Distplot:**

````python
# Extraemos las columnas que queremos graficar
df_displot = df_grades[["math score","reading score","writing score"]]

# Graficamos el displot indicando que no se genere el histograma ni rug plot, solamente las curvas
fig = ff.create_distplot([df_displot[c] for c in df_displot.columns], df_displot.columns, show_hist=False, show_rug=False)
fig.show()
````

**Histograma con Boxplots:**

````python
# Graficamos el histograma con boxplots, para esto se definió el parámetro marginal como "box", se agruparon por asignatura con
# el parámetro barmode, se modificó el ancho de las barras mediante el número de bins nbins,
fig = px.histogram(df_melt, x="value", color="subject", marginal="box", barmode='group', nbins=20, labels={'value':'grade'})
fig.show()
````

**Histograma con Faceta:**

````python
# Graficamos el histograma con faceta de filas definiendo un histograma por asignatura
fig = px.histogram(df_melt, x="value", color="subject", facet_row="subject",
                   category_orders={"subject": ["math score", "reading score", "writing score"]},labels={'value':'grade'}, nbins=15)
fig.show()
````

### Tabla de Resumen

Para generar la tabla de resumen:

- [X] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [X] Hacer una simplificación a través de un mapeo (investigar el método `map()`) de la variable `parental level of education`  según la siguiente conversión:

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [X] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas.
- [X] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [X] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'.


Utilizar la tabla de resultados esperados como guía para desarrollar este punto.

### 1.3.4 Información por Categorías [1 punto]

````python
# Copiamos los datos para la tabla de resumen
summary_table = df_grades.copy()

# Generamos un Dataframe con el cálculo de los promedios con cada asignatura para cada registro
means = summary_table[['math score','reading score','writing score']].mean(axis=1)

# Anexamos el Dataframe con los promedios como una nueva columna llamana 'GPA'
summary_table['GPA'] = means

summary_table.head()
````

````python
# Definimos la conversión para realizar el mapeo
mappings = {"some high school":"school","some college":"school", "high school":"school",
            "bachelor's degree":"college", "associate's degree":"college", "master's degree":"postgraduate"}

# Realizamos el mapeo sobre la variable 'parental level of education'
summary_table['simple parental level of education'] = summary_table['parental level of education'].map(mappings)

summary_table.head()
````

````python
# Se agrupa el Dataframe con respecto a las variables 'race/ethnicity' y 'simple parental level of education', además se calcula las medias
# de las variables numéricas
summary_table_mean = summary_table.groupby(['race/ethnicity','simple parental level of education']).mean(numeric_only=True).round(2)

summary_table_mean
````

````python
# Se agrupa el Dataframe con respecto a las variables 'race/ethnicity' y 'simple parental level of education', además se cuenta la cantidad
# de estudiantes en cada grupo
summary_table_count = summary_table.groupby(['race/ethnicity','simple parental level of education']).size().reset_index(name='count')

summary_table_count
````

````python
# Agregamos la columna 'count' al Dataframe con las medias de las notas mediante un merge entre ambos Dataframes
summary_table_final = summary_table_mean.reset_index().merge(summary_table_count, on=['race/ethnicity', 'simple parental level of education'])

summary_table_final.head()
````

````python
# Calculamos el total de estudiantes
total_alumnos = summary_table_final['count'].sum()

# Calculamos el porcentaje de alumnos con respecto al total y lo agregamos al Dataframe como una nueva columna
summary_table_final['percentage'] = (summary_table_final['count'] / total_alumnos * 100).round(2).astype(str) + '%'
````

````python
# Reorganizamos los índices para que el Dataframe quede agrupado con respecto a las variables 'race/ethnicity' y 'simple parental level of education'
summary_table_final.set_index(['race/ethnicity', 'simple parental level of education'], inplace=True)
summary_table_final
````

````python
pd.merge(df_grades,
         df_others,
         left_on="names",
         right_on="names",
         how="outer",)
````

> **Respuesta**: Vemos que la cantidad de filas aumenta, lo que indica que se están agregando alumnos posiblemente duplicados (hereda el número de filas del dataframe no limpiado).

### Left join

````python
pd.merge(df_grades,
         df_others,
         left_on="names",
         right_on="names",
         how="left",)
````

> **Respuesta**: Hereda la cantidad de filas del dataframe limpiado df_grades (dataframe de la izquierda), por lo que podemos inferir que no se heredan los datos corruptos del dataframe de la derecha

### Right join

````python
pd.merge(df_grades,
         df_others,
         left_on="names",
         right_on="names",
         how="right",)
````

> **Respuesta**: El resultado es similar que el outer join. La cantidad de filas aumenta (en este caso hereda las filas del dataframe de la derecha), lo que indica que se están agregando alumnos posiblemente duplicados (hereda el número de filas del dataframe no limpiado).

### Inner join

````python
pd.merge(df_grades,
         df_others,
         left_on="names",
         right_on="names",
         how="inner",)
````

> **Respuesta**: El resultado es similar que con el left join. Hereda la cantidad de filas del dataframe limpiado df_grades (dataframe de la izquierda), por lo que podemos inferir que no se heredan los datos corruptos del dataframe de la derecha.
